### 1. Write a Python program to find those numbers which are divisible by 7 and multiple of 5, between 1500 and 2700 (both included).

```python
def find_numbers(lower_bound, upper_bound):
    results=[]
    for i in range(lower_bound,upper_bound):
         if (i % 7 == 0 and i % 5 !=0):
                results.append(i)
                return results
lower_bound = int(input("Lower bound to search for numbers: ") )
upper_bound = int( input("Upper bound to search for numbers: ") )
found_numbers = find_numbers(lower_bound, upper_bound)
print("The numbers that are divisible by 7 but not by 5 are:\n{}".format(found_numbers))
```

```python
#### 4. Python Program for factorial of a number
num = int(input("Enter a number: "))
factorial = 1
if num < 0:
    print(" Factorial does not exist for negative numbers")
elif num == 0:
    print("The factorial of 0 is 1")
else:
    for i in range(1,num + 1):
        factorial = factorial*i
        print("The factorial of",num,"is",factorial)
```

```python
#### 5. Python Program for simple interest
def simple_interest(p,t,r):
      print('The principal is', p)
      print('The time period is', t)
      print('The rate of interest is',r)
      si = (p * t * r)/100
      print('The Simple Interest is', si)
      return si
simple_interest(10, 6, 8)
```

```python
####6. Python Program for compound interest
def compound_interest(principle, rate, time):
# Calculates compound interest
       Amount = principle * (pow((1 + rate / 100), time))
       CI = Amount - principle
       print("Compound interest is", CI)
compound_interest(10000, 10.25, 5)
```

```python
#### 7. Python Program to Check Armstrong Number
num = int(input("Enter a number: "))
sum = 0
temp = num
while temp > 0:
      digit = temp % 10
      sum += digit ** 3
      temp //= 10
        # display the result
      if num == sum:
            print(num,"is an Armstrong number")
      else:
          print(num,"is not an Armstrong number")
```

```python
####9. Python program to print all Prime numbers in an Interval
lower = 900
upper = 1000
print("Prime numbers between", lower, "and", upper, "are:")
for num in range(lower, upper + 1):
# all prime numbers are greater than 1
        if num > 1:
            for i in range(2, num):
                if (num % i) == 0:
                    break
                else:
                    print(num)
```
